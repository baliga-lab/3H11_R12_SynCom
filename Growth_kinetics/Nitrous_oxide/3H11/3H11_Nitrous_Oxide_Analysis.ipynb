```python
data = pd.read_excel('3H11-Nitrous-Oxide-Yield.xlsx',index_col=0)
data
```

```python
data['delta']=data['OD2']-data['OD1']
data
```

```python
temp = data[data.OD2<.3].dropna()
idx=[i for i in temp.index if '4' not in i]
temp=temp.loc[idx]
for idx, df in temp.groupby(by='Media'):
    plt.figure(figsize=(4,4))
    sns.boxplot(y='delta',hue='Nitrous Oxide',x='Nitrate',data=df)
    #sns.swarmplot(y='delta',hue='Nitrous Oxide',x='Nitrate',data=df)
    plt.title(idx)
    plt.ylabel('$ \Delta $ OD')
    plt.legend(bbox_to_anchor=(1,1),title='Nitrous Oxide')
    plt.ylim(-.01,.14)
```

```python
data = pd.read_csv('3H11-NitrousOxideYield-T0.csv').dropna()
data
```

```python
cond = {'A':'1 mL N2O','B':' 0.75 mL N2O','C':'0.5 mL N2O','D': '0.25 mL N2O','E':'0.1 mL N2O','F':'0 mL N2O','G':'Nitrate control','H':'N2O only control'}
for i in data.index:
    smp =data.loc[i,'Sample Name'][0]
    data.loc[i,'condition']=cond[smp]
data
```

```python
means=data[['Nitrous Oxide','condition']].groupby(by='condition').mean().reset_index()
means
```

```python
sns.barplot(x='condition',y='Nitrous Oxide',data=data)
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Nitrous Oxide (ppm)')
```
